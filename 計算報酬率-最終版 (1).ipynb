```python
#讀入淨值數據
data = pd.read_csv('/Users/weihongjie/python/淨值到1115.csv')
```

```python
#計算報酬率的函式
def calcRate(first,last):
    return (last-first)/first
```

```python
#導入刪除10檔後 490檔基金的名字
fundName = pd.read_csv('/Users/weihongjie/python/排除10檔的分群.csv')
```

```python
#讓date有pandas特有的格式 方便後續運算
data['newdate'] = pd.to_datetime(data['mdate'])
```

```python
#讓data只有需要的欄位(減少大小才能讓運算更快)
data= data[['coid','newdate','fld003']]
```

```python
#tf : 淨值的最後一天
#t0 : 淨值的第一天(星期一)
#讓t0每次加上7天，當t0加到等於tf
#需要計算特定日期的報酬率 改t0及tf就可以了
tf = pd.to_datetime('2018-11-16')
df = []
for funds in fundName.coid.unique():
    t0 = pd.to_datetime('2018-11-05')
    #星期一加上六天變星期日
    t1 = t0 + pd.Timedelta(days = 6)
    df.append(funds)
    while(t0<tf):
        temp = data[data['coid']==funds][data['newdate']>=t0][data['newdate']<=t1]
        print(temp)
        if(temp.index[0] == temp.index[-1]):
            df.append(np.nan) #若一個星期只有一天開盤，則補空白
        else:
#計算該星期開盤資料的最後一天(index為[-1])，和第一天(index為[0])的報酬率)
            df.append(calcRate(temp['fld003'].iloc[0],temp['fld003'].iloc[-1]))
        #print(temp)
        #print(temp['fld003'].iloc[0],temp['fld003'].iloc[-1])
        t0 = t0 +pd.Timedelta(weeks=1)
        t1 = t1 +pd.Timedelta(weeks=1)
```

```python
#將df做成DataFrame
newdf = pd.DataFrame()
newdf = newdf.append(df)
```

```python
#轉為numpy 方便 reshape
np1 = newdf.as_matrix()
np2 = np1.reshape(489,3)
```

```python
#將reshape後的np轉為DataFrame
newdf1 = pd.DataFrame(np2)
```

```python
#導出
newdf1.to_csv('最新報酬率.csv')
```
